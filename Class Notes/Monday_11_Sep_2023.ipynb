#**Python Lambda Function**

```python
x = lambda a: a + 10
print(x(10))
```

```python
x = lambda a, b: a * b + 10
print(x(2,8))
```

```python
x = lambda a, b, c: a * b * c
print(x(3, 6, 9))
```

```python
x = lambda x: "Even" if x % 2 ==0  else 'Not Even'

x = int(input)
print(x(2))
```

```python
result = lambda x,y : f"{x} is smaller than {y}" if x < y else (f"{x} is greater than {y}" if x > y  else f"{x} is equal to {y}")

print(result(12, 11))
print(result(12, 12))
print(result(12, 13))
```

```python
Max = lambda a, b : a if(a > b) else b

print(Max(12,121))
```

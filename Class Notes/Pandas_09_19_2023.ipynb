```python
dic_of_mean_wage = {}

for k in df.Education.value_counts().keys():
  dic_of_mean_wage[k] = df.Wage[df['Education'] == k].mean()
print(dic_of_mean_wage)
```

```python
max_key = None
max_value = float('-inf')
for key, value in enumerate(dic_of_mean_wage):
  if value > max_value:
        max_value = value
        max_key = key

print(max_key)
```

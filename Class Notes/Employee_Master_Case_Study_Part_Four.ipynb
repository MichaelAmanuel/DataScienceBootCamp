```python


class Employee:

    def __init__(self, emp_id, name, position, salary, age):
        self.emp_id = emp_id
        self.name = name
        self.position = position
        self.salary = salary
        self.age = age

    def display_info(self):
        print("Employee ID", self.emp_id)
        print("Name", self.name)
        print("Position", self.position)
        print("Salary", self.salary)
        print("Age", self.age)


class EmployeeManagementSystem:

    def __init__(self):
        self.employees = []

    def add_employee(self, emp):
        self.employees.append(emp)

    def remove_employee(self, emp_id):
        for emp in self.employees:
            if emp.emp_id == emp_id:
                self.employees.remove(emp)
                return True
        return False

    def update_name(self, emp_id, name):
        for emp in self.employees:
            if emp.id == emp_id:
                self.employees.name = name
                return True
        return False

    def update_salary(self, emp_id, salary):
        for emp in self.employees:
            if emp.id == emp_id:
                self.employees.salary = salary
                return True
        return False

    def update_age(self, emp_id, age):
        for emp in self.employees:
            if emp.emp_id == emp_id:
                self.employees.age = age
                return True
        return False

    def delete_by_name(self, name):
        all_matching_names = []
        for employee in self.employees:
            if employee.name == name:
                all_matching_names.append(employee)
        if len(all_matching_names) == 0:
            return False
        elif len(all_matching_names) == 1:
            self.employees.remove(all_matching_names[0])
            return True
        elif len(all_matching_names) > 1:
            print("Duplicate names found please select the id you want to remove")
            all_ids = []
            for emps in all_matching_names:
                print('ID :-', emps.emp_id)
                all_ids.append(emps.emp_id)
            id_to_remove = int(input("Please choose the id you want to remove :- "))
            if id_to_remove in all_ids:
                self.employees.remove(all_matching_names[all_ids.index(id_to_remove)])
                return True
            else:
                return False

    def details_by_id(self, emp_id):
        for employee in self.employees:
            if employee.emp_id == emp_id:
                employee.display_info()

    def details_by_name(self, name):
        for employee in self.employees:
            if employee.name == name:
                employee.display_info()

    def details_all_employees(self):
        print("Check length", len(self.employees))
        for employee in self.employees:
            employee.display_info()

            print("====================================")

    def write_data(self):
        with open("employee_data.txt", "a") as f:
            for employee in self.employees:
                f.write(str(employee.name) + ", " + str(employee.emp_id)+\
                        ", "+str(employee.position)+", "+str(employee.salary)+\
                        ", "+str(employee.age)+"\n")


emp_managed_system = EmployeeManagementSystem()

with open("employee_data.txt", "r") as f:
    data = f.read()
    all_objects = data.split('\n')

    for employee in all_objects:
      employee = employee.split(', ')
      if len(employee) > 1:
        name = employee[0]
        emp_id = employee[1]
        position = employee[2]
        salary = employee[3]
        age = employee[4]
        emp_temp = Employee(name, emp_id, position, salary, age)



emp_system = True

while emp_system:
    print("Please choose the option.  ")
    print("1) Add Employee. ")
    print("2) Delete Employee by ID. ")
    print("3) Delete Employee by Name. ")
    print("4) All details.")
    print("5) Exit.")

    choice = input("Please choose an option. ")
    if choice == '1':
        name_choice = input("Please enter Employee name. ")
        age_choice = input("Please enter Employee age. ")
        salary_choice = input("Please enter Employee salary. ")
        id_choice = int(input("Please enter Employee ID. "))
        position_choice = input("Please enter Employee position. ")

        id_user = len(emp_managed_system.employees)
        emp_temp = Employee(id_choice, name_choice, position_choice, salary_choice, age_choice)
        emp_managed_system.add_employee(emp_temp)

    elif choice == '2':
        id_user = int(input("Please enter Employee id you want to remove "))
        if emp_managed_system.remove_employee(id_user):
            print("Employee Removed Successfully.")
        else:
            print("Id not found could not remove the employee")

    elif choice == '3':
        name_input = input('Please enter Employee name you want to remove.')
        if emp_managed_system.delete_by_name(name_input):
            print('Employee Removed Successfully.')
        else:
            print("Id not found could not remove the employee")

    elif choice == '4':
        emp_managed_system.details_all_employees()

    elif choice == '5':
        print('Bye!')
        emp_managed_system.write_data()

        emp_system = False

```

```python
# lambda

x = lambda x, y: x * y
x(12, 23)
```

```python
str1 = 'Python Programming'
x = lambda str1: str1.upper()
x(str1)
```

```python
x = lambda a, b: a if (a > b) else b
x(12, 23)
```

```python
numbers = [1, 2, 3, 4, 5]
doubled_numbers = list(map(lambda x: x * 2, numbers))
print(doubled_numbers)
```

```python
numbers = [1, 2, 3, 4, 5]
doubled_numbers = lambda x: x * 2, numbers
doubled_numbers
```

```python
# *args, **kwargs

def func_one(a, b):
  return a * b

func_one(12, 23)
```

```python
def func_two(*args):
  sum = 0
  for i, num in enumerate(args):
    sum = sum + num
  return sum


func_two(12, 23, 34, 45, 56, 67, 78, 89, 90)
```

```python
def func_3(**kwargs):
  return kwargs
```

```python
func_3( a='apple', b='banana', c='cherries', d='dates', e='eggfriut')
```

```python
def func_4(a, b, *args, **kwargs):
  return a, b, args, kwargs
```

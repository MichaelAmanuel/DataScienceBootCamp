```python
class Employee:

    def __init__(self, emp_id, name, salary,  age, job_position):
        self.name = name
        self.salary = salary
        self.id = emp_id
        self.age = age
        self.job = job_position

    def display_info(self):
        print("Employee ID", self.id)
        print("Name is", self.name)
        print('Age is ', self.age)
        print('Salary', self.salary)
        print('')


class EmployeeManagementSystem:
    def __init__(self):
        self.all_employees_list = []

    def add_employee(self, employee_object):
        self.all_employees_list.append(employee_object)

    def delete_by_id(self, the_id_to_delete):
        remove = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_delete:
                self.all_employees_list.remove(employee)
                remove = True
                break
        if remove:
            print("Employee removed Successfully.")
        else:
            print('No Such Employee found.')

    def update_name_by_id(self,the_id_to_update, the_name_to_update):
        update = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_update:
                employee.name = the_name_to_update
                update = True
        if update:
            print("Employee name Successfully update.")
        else:
            print("No Such Employee found.")

    def update_salary_by_id(self, the_id_to_update, the_salary_to_update):
        update = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_update:
                employee.salary = the_salary_to_update
                update = True
        if update:
            print("Employee salary Successfully update.")
        else:
            print("No Such Employee found.")

    def update_age_by_id(self, the_id_to_update, the_age_to_update):
        update = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_update:
                employee.salary = the_age_to_update
                update = True
        if update:
            print("Employee Age Successfully update.")
        else:
            print("No Such Employee found.")

    def get_details_by_id(self, the_id_for_details):
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_for_details:
                employee.display_info()

    def get_details_by_name(self, the_name_for_details):
        all_employee_with_same_name = []
        for employee in self.all_employees_list:
            if employee.emp_name == the_name_for_details:
                all_employee_with_same_name.append(employee)

        if len(all_employee_with_same_name) == 0:
            print("No Such Employee Found in Record")
        elif len(all_employee_with_same_name) == 1:
            all_employee_with_same_name[0].display_info()
        elif len(all_employee_with_same_name) > 1:
            print("Found multiple employee with same name please choose an ID you would like to see detail of ")
            id_of_the_name_details = int(input("Please enter the ID. "))
            for employee in all_employee_with_same_name:
                if employee.emp_id == id_of_the_name_details:
                    employee.display_info()

    def delete_by_name(self, the_name_to_remove):
        all_employee_with_same_name = []
        for employee in self.all_employees_list:
            if employee.emp_name == the_name_to_remove:
                all_employee_with_same_name.append(employee)
        if len(all_employee_with_same_name) == 0:
            print("No Such Employee Found in Record")
        elif len(all_employee_with_same_name) == 1:
            self.all_employees_list.remove(all_employee_with_same_name[0])
            print("Record removed Successfully. ")
        else:
            print("Found multiple employee with same name please choose an ID you would like to see detail of ")
            id_of_the_name_details = int(input("Please enter the ID. "))
            for employee in all_employee_with_same_name:
                if employee.emp_id == id_of_the_name_details:
                    employee.get_details_by_id(id_of_the_name_details)

    def print_all_employee_details(self):
        for employee in self.all_employees_list:
            employee.display_info()
            print("=====================================")


controller_class_object = EmployeeManagementSystem()

emp_system = True

while emp_system:
    print("Please choose the option.  ")
    print("1) Add Employee. ")
    print("2) Delete Employee by ID. ")
    print("3) Delete Employee by Name. ")
    print("4) All details.")
    print("5) Update name by ID. ")
    print("6) Update salary by ID. ")
    print("7) Update Age by ID. ")
    print("8) Update Position by ID. ")
    print('press any other if you want to exit employee management system. ')

    choice = input("Please choose an option. ")
    if choice == '1':
        name_choice = input("Please enter Employee name. ")
        age_choice = input("Please enter Employee age. ")
        salary_choice = input("Please enter Employee salary. ")
        id_choice = int(input("Please enter Employee ID. "))
        position_choice = input("Please enter Employee position. ")

        id_user = len(controller_class_object.all_employees_list)
        emp_object = Employee(id_choice, name_choice, position_choice, salary_choice, age_choice)
        controller_class_object.add_employee(emp_object)

    elif choice == '2':
        id_user = int(input("Please enter Employee id you want to remove "))
        if controller_class_object.delete_by_id(id_user):
            print("Employee Removed Successfully.")
        else:
            print("Id not found could not remove the employee")

    elif choice == '3':
        name_input = input('Please enter Employee name you want to remove.')
        if controller_class_object.delete_by_name(name_input):
            print('Employee Removed Successfully.')
        else:
            print("Id not found could not remove the employee")

    elif choice == '4':
        controller_class_object.print_all_employee_details()

    elif choice == '5':
        id_to_update = int(input("Please enter Employee Is you want to Update. "))
        name_to_update = input("Please enter Name want to Update the Em  ")
        controller_class_object.update_name_by_id(id_to_update, name_to_update)

    elif choice == '6':
        id_to_update = int(input("Please enter Employee Is you want to Update. "))
        salary_to_update = input("Please enter Name want to Update the Em  ")
        controller_class_object.update_salary_by_id(id_to_update, salary_to_update)

    elif choice == '7':
        id_to_update = int(input("Please enter Employee Is you want to Update. "))
        age_to_update = input("Please enter Name want to Update the Em  ")
        controller_class_object.update_age_by_id(id_to_update, age_to_update)

    elif choice == '8':
        id_to_update = int(input("Please enter Employee Is you want to Update. "))
        position_to_update = input("Please enter Name want to Update the Em  ")
        controller_class_object.update_age_by_id(id_to_update, position_to_update)

    elif choice == '9':
        id_to_print_details = int(input("Please enter Employee Id you want to details. "))
        controller_class_object.get_details_by_id(id_to_print_details)

    elif choice == '10':
        name_to_print_details = int(input("Please enter Employee name you want to detail. "))
        controller_class_object.get_details_by_name(name_to_print_details)

    elif choice == '11':
        print("Bye !!!!!!")
        emp_system = False

```

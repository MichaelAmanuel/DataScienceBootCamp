```python
class Computer:

  def __init__(self, name, ram, rom, cpu, harddisk, memory):
    self.com_name = name
    self.ram = ram
    self.rom = rom
    self. cpu = cpu
    self.harddisk = harddisk
    self.memory = memory


  def type_pc(self):
    print(f"The computer name is {self.com_name}.")
    print(f'The computer ram {self.ram}')
    print(f'The computer rom {self.rom}')
    print(f'The computer cpu {self.cpu}')
    print(f'The computer harddisk: {self.harddisk}')
    print(f'The computer memory {self.memory}')

pc_1 = Computer('ThinkPad','8Gb', 24, 'i7', '250Gb', 'disk')
pc_1.type_pc()
```

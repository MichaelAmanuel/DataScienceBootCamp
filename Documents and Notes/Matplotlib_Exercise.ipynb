```python
import numpy as np
import pandas as pd
```

```python
x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.2,0.5,0.2,0.2])

ax.plot(x,y, color='r')
ax1.plot(x,y, color='r')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.2,0.5,0.2,0.2])

ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax1.plot(x,y)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')


ax1.set_xlim([20,22])
ax1.set_ylim([30,50])
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols = 2, dpi=100)

axes[0].plot(x,y, color='blue', linestyle='--')
axes[1].plot(x,z, color='r')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from standard normal distribution
standard_normal_data = np.random.randn(1000)

# Plot histogram to visualize the distribution
plt.hist(standard_normal_data, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from uniform distribution between 0 and 1
uniform_data = np.random.rand(1000)

# Plot histogram to visualize the distribution
plt.hist(uniform_data, bins=30, density=True, alpha=0.7, color='green')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
# Create a DataFrame with a MultiIndex
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1, 2, 3, 1, 2, 3]
hier_index = list(zip(outside, inside))
multi_index = pd.MultiIndex.from_tuples(hier_index)

df = pd.DataFrame(data={'Values': [10, 20, 30, 40, 50, 60]}, index=multi_index)

# Display the DataFrame
print("DataFrame:")
print(df)

# Get the names of index levels
index_names = df.index.names
print("\nIndex Names:", index_names)
```

```python
import pandas as pd

data = {'Company': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
        'Person': ['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
        'Sales': [200, 120, 340, 124, 243, 350]}

df = pd.DataFrame(data)

# Grouping by 'Company' and calculating the total sales for each company
grouped_df = df.groupby('Company')
# Displaying the result
print(grouped_df)
```

```python
grouped_df.mean()
```

```python
grouped_df.sum()
```

```python
grouped_df.sum().loc['FB']
```

```python
df.groupby('Company').sum().loc['FB']
```

```python
df.groupby('Company').describe().transpose()
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
merged_df = pd.merge(left, right, on='key')
print("Merged DataFrame:")
print(merged_df)
print()
concatenated_df = pd.concat([left, right], axis=1)
print("Concatenated DataFrame:")
print(concatenated_df)

```

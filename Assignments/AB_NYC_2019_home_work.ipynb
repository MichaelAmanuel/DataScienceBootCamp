```python
# 1) Calculate How many properties in each neighbourhood group

dataset.groupby('neighbourhood_group')['id'].sum()
```

```python
# 2) Which Neighbour Hood Group Has highest prices use bar chart
#  to show this for all neighbourhood groups

max_out = dataset.groupby('neighbourhood_group')['price'].mean().reset_index()
max_data = max_out.sort_values(by='price', ascending=False)
sns.barplot(x='neighbourhood_group', y='price', data=max_data)
```

```python
# 3) Which Neighbourhood group has lowest prices use
#  bar chart to show this for all neighbourhood groups

min_out = dataset.groupby('neighbourhood_group')['price'].mean().reset_index()
min_data = max_out.sort_values(by='price', ascending=True)
sns.barplot(x='neighbourhood_group', y='price', data=min_data)
```

```python
# 4) Highest and lowest price airbnb in each Neighbour Hood Group



grouped = dataset.groupby('neighbourhood_group')


highest_prices = pd.DataFrame()
lowest_prices = pd.DataFrame()


for group_name, group_data in grouped:

    highest_price_row = group_data[group_data['price'] == group_data['price'].max()]
    lowest_price_row = group_data[group_data['price'] == group_data['price'].min()]

    highest_prices = highest_prices.append(highest_price_row)
    lowest_prices = lowest_prices.append(lowest_price_row)

highest_prices.reset_index(drop=True, inplace=True)
lowest_prices.reset_index(drop=True, inplace=True)

print("Highest Price Listings:")
print(highest_prices)

print("\nLowest Price Listings:")
print(lowest_prices)
```

```python
# 5) Which Host has highest count of properties

host_property_counts = dataset['host_id'].value_counts()

highest_count_host_id = host_property_counts.idxmax()
highest_count_properties = host_property_counts.max()

highest_count_host_name = dataset[dataset['host_id'] == highest_count_host_id]['host_name'].iloc[0]

print(f"The host with the highest count of properties is '{highest_count_host_name}' (host_id: {highest_count_host_id}) with {highest_count_properties} properties.")
```

```python
# 6) Which neighbourhood is most expensive and cheapest top 10 barchart


neighborhood_prices = dataset.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)

top_10_expensive = neighborhood_prices.head(10)
top_10_cheapest = neighborhood_prices.tail(10)

plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
top_10_expensive.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Expensive Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Average Price')


plt.subplot(1, 2, 2)
top_10_cheapest.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Cheapest Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Average Price')

# Adjust spacing between subplots
plt.tight_layout()

# Show the bar chart
plt.show()
```

```python
# 7) Count different type of room_type and do a pie chart
# showing % of each room type in overall datase

room_type_counts = dataset['room_type'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Room Types')
plt.axis('equal')
plt.show()

```

```python
# 8) scatter plot between longitutude and latitude
# for different neighbourhood groups

neighborhood_groups = dataset['neighbourhood_group'].unique()

plt.figure(figsize=(12, 8))

for group in neighborhood_groups:
    group_data = dataset[dataset['neighbourhood_group'] == group]
    plt.scatter(group_data['longitude'], group_data['latitude'], label=group, alpha=0.5)

plt.title('Scatter Plot of Longitude vs. Latitude for Different Neighborhood Groups')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)

plt.show()
```

```python

room_types = dataset['room_type'].unique()

plt.figure(figsize=(12, 8))

for room_type in room_types:
    room_type_data = dataset[dataset['room_type'] == room_type]
    plt.scatter(room_type_data['longitude'], room_type_data['latitude'], label=room_type, alpha=0.5)

plt.title('Scatter Plot of Longitude vs. Latitude for Different Room Types')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)

plt.show()
```

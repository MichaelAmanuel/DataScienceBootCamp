```python
#a-1. Construct a contingency table that cross-classifies the data by
# Sex and Feasible. Provide the frequencies in the accompanying table.


contingency_table = pd.crosstab(df['Sex'], df['Feasible'])
```

```python
contingency_table
```

```python
# a-2. How many of the students were female?


contingency_table.loc['female'].sum()
```

```python
contingency_table.loc['male'].sum()
```

```python
#  a-3. How many of the students felt that it was feasible for
#  men and women to be just friends?

contingency_table['yes'].sum()
```

```python
df_cleaned['Feasible'].value_counts().get('yes', 0)
```

```python
# b-1. What is the likelihood that a male student
# feels that men and women can be just friends?
# (Report the proportion rounded to 2 decimal places.)

contingency_table['yes']['male'].sum()
```

```python
contingency_table['yes']['female'].sum()
```

```python
# b-2. What is the likelihood that a female student
# feels that men and women can be just friends?
 # (Report the proportion rounded to 2 decimal places.)

female_yes_count = contingency_table.loc['female']['yes']

total_female_count = contingency_table.loc['female'].sum()

likelihood_female_yes = female_yes_count / total_female_count
```

```python
round(likelihood_female_yes, 2)
```

# 3)  Find Total Sale between two dates entered by user

```python

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%y %H:%M')


start_date_input = input("Enter the start date (YYYY-MM-DD): ")
end_date_input = input("Enter the end date (YYYY-MM-DD): ")

start_date = pd.to_datetime(start_date_input)
end_date = pd.to_datetime(end_date_input)


filtered_df = df[(df['InvoiceDate'] >= start_date) & (df['InvoiceDate'] <= end_date)]


total_sale = filtered_df['Quantity'].sum()

print(f"Total sale between {start_date} and {end_date}: ${total_sale:.2f}")


```

```python
# 8) Find how many different customers are available
#  for each country save the count in dictionary also
# return countries with highest and lowest customer count

customer_counts = {}
for country, group in df_sorted.groupby('Country'):
    customer_counts[country] = group['CustomerID'].nunique()

max(customer_counts, key=customer_counts.get)
```

```python
# 8) Find how many different customers are available
#  for each country save the count in dictionary also
# return countries with highest and lowest customer count

customer_counts = {}
for country, group in df_sorted.groupby('Country'):
    customer_counts[country] = group['CustomerID'].nunique()

max(customer_counts, key=customer_counts.get)
```

```python
customer_counts = {}
for country, group in df_sorted.groupby('Country'):
    customer_counts[country] = group['CustomerID'].nunique()

min(customer_counts, key=customer_counts.get)
```

```python
# 9) Find which stock code was ordered the most by each country

stock_by_country = df_sorted.groupby(['Country', 'StockCode']).size().reset_index(name='Count')
stock_by_country = stock_by_country.sort_values(by='Count', ascending=False)
stock_by_country.drop_duplicates('Country')
```
